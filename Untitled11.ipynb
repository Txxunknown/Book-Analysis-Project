```python
#column isbn should contain only ineger value and remove null and also conert it into integer


df=df[df['isbn'].notnull()]

df['isbn'] = df['isbn'].str.replace(r'\D', '', regex=True)
df = df[df['isbn'].str.len() > 0]

df['isbn']=df['isbn'].astype('int')
```
